```python
with open('coca-macm/coca20k-macm.html') as file:
    data = file.read()
```

```python
lst = data.split('\n            \n             <tr>')[1:]
```

```python
def find_phonetic(data):
    bsObj = BeautifulSoup(data, "lxml")
#     mylist1 = bsObj.find_all('font', {'color':'#ff5400'})
    mylist2 = bsObj.find_all('font', {'color':'#21887d'})
    
#     lst1 = [i.get_text() for i in mylist1]
    lst2 = [i.get_text() for i in mylist2]
    return '\n'.join(lst2)
```

```python
df_html = pd.DataFrame({'all': lst})
```

```python
df_html.head()
```

```python
df_csv = pd.read_csv('coca-macm/coca20k-macm.csv').ix[:, 1:]
```

```python
df_csv.head()
```

```python
df_csv.rename(columns={
        '单词': 'words',
        '解释': 'des',
    }, inplace=True)
```

```python
df = pd.concat([df_csv.ix[:, ['words']], df_html], axis=1 )
```

```python
df['all_speech'] = df['all'].map(find_phonetic)
```
